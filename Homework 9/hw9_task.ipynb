```python
from sklearn.datasets import make_blobs, make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MaxAbsScaler
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib
import copy
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
def read_data(path="../datasets/boston.csv"):
    dataframe = np.genfromtxt(path, delimiter=",", skip_header=15)
    X = dataframe[:, :-1]
    y = dataframe[:, -1]
    return X, y
```

```python
def generate_synthetic(size, dim=6, noise=0.1):
    X = np.random.randn(size, dim)
    w = np.random.randn(dim + 1)
    noise = noise * np.random.randn(size)
    y = X.dot(w[1:]) + w[0] + noise
    return X, y
```

```python
def mse(y_true, y_predicted):
    return np.mean((y_true - y_predicted)**2)
```

### Задание 2 (3 балла)
Теперь реализуем линейную регрессию при помощи явного решения задачи минимизации. 

#### Методы
`fit(X, y)` - решает задачу минимизации $\arg\min_{w, b}\sum ((w\cdot x + b) - y)^2$. 

`predict(X)` - строит предсказание `y` для объектов из `X`.

```python
class NormalLR:
    def __init__(self):
        self.w = None
        
    def fit(self, X, y):
        self.w = np.linalg.inv(X.T @ X) @ X.T @ y
        
    def predict(self, X):
        return X @ self.w
```

```python
def add_ones(X):
    return np.c_[X, np.array([1] * X.shape[0])]
```

```python
X, y = generate_synthetic(1024, dim=6)
X = add_ones(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
regr = NormalLR()
regr.fit(X_train, y_train)
print(mse(y_test, regr.predict(X_test)))
```

### Задание 4 (2 балла)
Добавьте в метод `fit` регуляризацию Лассо с коэффициентом `l`. Постройте график зависимости ошибки предсказания данных из синтетического набора данных от коэффициента регуляризации.

```python
class GradientLR:
    def __init__(self, alpha, iterations=10000, l=0):
        self.learning_rate = alpha
        self.iterations = iterations
        self.l = l
        self.w = None
    def fit(self, X, y):
        self.w = np.array([1] * X.shape[1])
        
        for i in range(self.iterations):
            grad = 2/X.shape[0] * X.T @ (X @ self.w - y)
            self.w = self.w - self.learning_rate * grad - self.l * np.sign(self.w)

    def predict(self, X):
        return X @ self.w
```

```python
def build_plot(X_train, y_train, X_test, y_test):
    x, y = np.arange(0, 1.5, 0.1), []
    for lasso in x:
        regr = GradientLR(0.002, iterations=1000, l=lasso)
        regr.fit(X_train, y_train)
        y.append(mse(y_test, regr.predict(X_test)))
    plt.figure(figsize=(15, 8))
    plt.plot(x, y)
    plt.xlabel('Lasso coef')
    plt.ylabel('Mean square error')
```

```python
X, y = generate_synthetic(1024, noise=0.1)
X = add_ones(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
regr = GradientLR(0.001, iterations=100000, l=0)
regr.fit(X_train, y_train)
print(mse(y_test, regr.predict(X_test)))
build_plot(X_train, y_train, X_test, y_test)
```
