```python
from sklearn.datasets import make_blobs, make_moons
from sklearn.model_selection import train_test_split
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib
import copy
```

```python
def read_data(path="../datasets/boston.csv"):
    dataframe = np.genfromtxt(path, delimiter=",", skip_header=15)
    X = dataframe[:, :-1]
    y = dataframe[:, -1]
    return X, y
```

```python
def generate_synthetic(size, dim=6, noise=0.1):
    X = np.random.randn(size, dim)
    w = np.random.randn(dim + 1)
    noise = noise * np.random.randn(size)
    y = X.dot(w[1:]) + w[0] + noise
    return X, y
```

### Задание 1 (1 балл)
Для начала нужно понять, какую метрику для ошибки будем использовать. В нашем случае нам подойдет стандартная метрика MSE. Ее и нужно реализовать.

```python
def mse(y_true, y_predicted):
    return np.mean((y_true - y_predicted)**2)
```

### Задание 2 (3 балла)
Теперь реализуем линейную регрессию при помощи явного решения задачи минимизации. 

#### Методы
`fit(X, y)` - решает задачу минимизации $\arg\min_{w, b}\sum ((w\cdot x + b) - y)^2$. 

`predict(X)` - строит предсказание `y` для объектов из `X`.

```python
class NormalLR:
    def __init__(self):
        self.a = None
        self.b = None
        
    def fit(self, X, y):
        sx, sy, sxy, sx_y, sxx, n = *self.calculate_features(X, y), len(y)
        
        self.a = (n * sxy - sx_y)/(n * sxx - sx ** 2)
        self.b = (sy - self.a @ sx)/n
        
    def predict(self, X):
        return X @ self.a + self.b
    
    def calculate_features(self, X, y):
        sx = X.sum(axis=0) # sum(x_i)
        sy = sum(y) # sum(y_i)
        sxy = np.array([x * y for x, y in zip(X, y)]).sum(axis=0) # sum(x_i * y_i)
        sx_y = sx * sy # sum(x_i) * sum(y_i)
        sxx = (X * X).sum(axis=0) # sum(x_i^2)
        return sx, sy, sxy, sx_y, sxx
```

```python
X, y = generate_synthetic(1024, dim=6)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
regr = NormalLR()
regr.fit(X_train, y_train)
print(mse(y_test, regr.predict(X_test)))
```

#### Небольшой пример, что все работает для меньших размерностей

Теперь реализуем линейную регрессию с использованием градиентного спуска с larning rate `alpha` в течении `iterations` итераций.

#### Методы
`fit(X, y)` - приближает решение задачи минимизации $\arg\min_{w, b}\sum ((w\cdot x + b) - y)^2$ при помощи градиентного спуска. 


`predict(X)` - строит предсказание `y` для объектов из `X`.

### Задание 4 (2 балла)
Добавьте в метод `fit` регуляризацию Лассо с коэффициентом `l`. Постройте график зависимости ошибки предсказания данных из синтетического набора данных от коэффициента регуляризации.

```python
class GradientLR:
    def __init__(self, alpha, iterations=10000, l=0.01):
        self.learning_rate = alpha
        self.iterations = iterations
        self.l = l
        self.a = None
        self.b = None
        
    def fit(self, X, y):
        self.a, self.b = np.array([1] * X.shape[1]), 0
        
        for i in range(self.iterations):
            error = self.predict(X) - y
            a_grad = (X * error.reshape(-1, 1)).mean(axis=0)
            b_grad = error.mean()
            self.a = self.a - self.learning_rate * a_grad - self.l * np.sign(self.a)
            self.b = self.b - self.learning_rate * b_grad - self.l * np.sign(self.b)
    
    def predict(self, X):
        return X @ self.a + self.b
```

```python
regr = GradientLR(alpha=0.001, iterations=5, l=0)
regr.fit(X_train, y_train)
print(mse(y_test, regr.predict(X_test)))
build_plot(X_train, y_train, X_test, y_test)
```

```python
def build_plot(X_train, y_train, X_test, y_test):
    x, y = np.arange(0, 10, 0.5), []
    for lasso in x:
        regr = GradientLR(0.01, iterations=500, l=lasso)
        regr.fit(X_train, y_train)
        y.append(mse(y_test, regr.predict(X_test)))
        print(sum(regr.a) + regr.b)
    plt.figure(figsize=(15, 8))
    plt.plot(x, y)
    plt.xlabel('Lasso coef')
    plt.ylabel('Mean square error')
```

```python
X, y = generate_synthetic(1024, noise=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
regr = GradientLR(0.01, iterations=10000, l=0)
regr.fit(X_train, y_train)
print(mse(y_test, regr.predict(X_test)))
build_plot(X_train, y_train, X_test, y_test)
```
